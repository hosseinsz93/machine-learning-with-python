```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot([1, 2, 3, 4])
```

```python
plt.plot([1, 2, 3, 4],[1, 2, 3, 4], 'ro')
plt.ylabel('some numbers')
```

```python
t = np.arange(0., 5., 0.2)
t
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.legend(['linear', 'quadratic', 'cubic'], loc='upper left')
```

```python
data = {
    'a': np.arange(500), # array from 0 to 499
    'b': np.random.randint(0, 50, 500), # 50 random integers between 0 and 500
    'c': np.random.randn(500) # 500 random numbers from standard normal distribution
}
data
```

```python
plt.scatter(data['a'], data['b'])  # scatter in useful for ML applications
```

```python
plt.scatter(data['a'], data['c'])
```
